```python
#Read profile (all) output
def view_all():
    Output.delete(0,END)
    x = ""
    for i in profile:
        for v in profile[i]:
            x += profile[i][v] + "   " 
        Output.insert(END,x)
        x = ""
    return

#Read profile(all) then search by value
def search_all():
    x=""
    name_text = Input.get().lower()
    Output.delete(0,END)
    
    if name_text == '':
        Output.delete(0,END)
        messagebox.showinfo("Error","Please Enter Name")
        return
    else:
        for i in profile:
            if name_text.lower() == profile[i]['NAME'][:len(name_text)].lower():
                for v in profile[i]:
                    x += profile[i][v] + "   "
                Output.delete(0,END) 
                Output.insert(END,x)
                x = ""
                check = 1
                break
            else:
                check = 0    
        if check == 1:
            pass
        else:
            Output.delete(0,END)
            messagebox.showinfo("Error","Please Enter Name Correctly")
        return

#take input from profile into Entrybox by cursorselection  
def get_selected_row(event):
    
    global selected_tuple
    Index = Output.curselection()
    selected_tuple = Output.get(Index)
    print(selected_tuple)
    
    Input.delete(0,END)
    text1.delete(0,END)
    text2.delete(0,END)
    text3.delete(0,END)
    text4.delete(0,END)
    text5.delete(0,END)
    #split data into each entry
    for i in profile:
        profileList = Output.get(Index).split("   ") 
        if profile[i]['ID'] == profileList[0]:
            student = profile[i]['NAME']
            textid = profile[i]['ID']
            texttel = profile[i]['TELNUMBER']
            textline = profile[i]['LINE_ID']
            textemail = profile[i]['EMAIL_KMUTNB']
            Input.insert(END,student)
            text1.insert(END,textid)
            text2.insert(END,student)
            text3.insert(END,texttel)
            text4.insert(END,textline)
            text5.insert(END,textemail)
    
    return(selected_tuple)
#Update into JSON
#Confirmation

def update_all():
    Index = Output.curselection()
    if Index == (0,) or Index == (1,) or Index == (2,):
        profileList = Output.get(Index).split("   ") 
        print(profileList)
        nametext=textname_box.get()
        idtext=textid_box.get()
        teltext=texttel_box.get()
        linetext=textline_box.get()
        emailtext=textemail_box.get()
        if nametext == "" or idtext == ""or teltext == "" or linetext == "" or emailtext == "" :
            checkupdate = messagebox.askquestion("Empty", "Some Entry is Empty, Do you want to Update?") 
            if checkupdate == "yes":
            #Write new value from entry into profile
                for i in profile:
                    print(i)
                    if  profile[i]["ID"] == profileList[0] :
                        profile[i]["ID"] = textid_box.get()
                        profile[i]["NAME"] = textname_box.get()
                        profile[i]["TELNUMBER"] = texttel_box.get()
                        profile[i]["LINE_ID"] = textline_box.get() 
                        profile[i]["EMAIL_KMUTNB"] = textemail_box.get()
                        info = 1
                        with open("profile.json", "w") as jsonFile: 
                            json.dump(profile,jsonFile,indent = 4)
                        view_all()
                        break
        else:
            checkupdate = messagebox.askquestion("Update", "Do you want to Update?") 
            if checkupdate == "yes":
            #Write new value from entry into profile
                for i in profile:
                    print(i)
                    if  profile[i]['ID'] == profileList[0] :
                        profile[i]["ID"] = textid_box.get()
                        profile[i]["NAME"] = textname_box.get()
                        profile[i]["TELNUMBER"] = texttel_box.get()
                        profile[i]["LINE_ID"] = textline_box.get() 
                        profile[i]["EMAIL_KMUTNB"] = textemail_box.get()
                        info = 1
                        with open("profile.json", "w") as jsonFile: 
                            json.dump(profile,jsonFile,indent = 4)
                        view_all()
                        break
                        
                if info == 1:
                    messagebox.showinfo("Success","Update Successful !")         
    else:
        messagebox.showinfo("Error","Please Select Line First")  
    return
    view_all()
#Close
def Exit():
    respone = messagebox.askquestion("Exit", "Do you want to Exit?") 
    if respone == "yes":
        root.destroy()
```
